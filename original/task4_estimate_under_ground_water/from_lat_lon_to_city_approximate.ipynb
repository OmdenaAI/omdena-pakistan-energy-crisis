## fuzzy join ng_geoloc with nasa_nigeria_weather_median_sun to translate lat, lon in the sunlight dataframe to have city or area name

## similarly fuzzy join ng_geoloc with bgs_nigeria_groundwater to translate lat, lon in the groundwater dataframe to have city or area name

https://rdrr.io/cran/fuzzyjoin/man/geo_join.html

```r
install.packages("fuzzyjoin")
```

```r
library("fuzzyjoin")
```

```r
solar_city <- geo_inner_join(ng_geoloc, nasa_nigeria_weather_median_sun, by = NULL, method = "haversine", max_dist = 22,
  )  #places with alt, lon 20 miles away getting join ... we are doing fuzzy join as the lat, lon do not match exactly
```

```r
solar_city
```

```r
nrow(solar_city)
```

```r
solar_city
```

```r
water_city <- geo_inner_join(ng_geoloc, bgs_nigeria_groundwater, by = NULL, method = "haversine", max_dist = 22,
  )  #places with alt, lon 20 miles away getting join ... we are doing fuzzy join as the lat, lon do not match exactly
```

```r
nrow(water_city)
```

```r
head(water_city)
```

```r
solar_water_combined <- merge(solar_city, water_city, by='city')
```

```r
tail(solar_water_combined)
```

```r
nrow(solar_water_combined)
```

```r
length(unique(solar_water_combined$city))
```

```r
calculate_mode <- function(x) {
  uniqx <- unique(x)
  uniqx[which.max(tabulate(match(x, uniqx)))]
}
```

```r
solar_water_combined%>%
group_by(city)%>%
summarise(mean_sun=mean(ALLSKY_SFC_SW_DWN), median_sun=median(ALLSKY_SFC_SW_DWN), 
          max_sun=max(ALLSKY_SFC_SW_DWN),
          min_sun=min(ALLSKY_SFC_SW_DWN),
          mode_water=calculate_mode(GWSTOR_V2))
```
