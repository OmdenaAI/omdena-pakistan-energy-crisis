```python
import rasterio
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
"""
convert index to lat-lon & vice versa. Because pop & nighttime use the same resolution & crs
so we do not need to concern about mismatch
"""
def convert_pixel_to_coord(px, py):
    with rasterio.open('worldpop_2019.tif') as map_layer:
        pixels2coords = map_layer.xy(px, py)  #input px, py
    return pixels2coords
```

```python
df = pd.read_csv('populated_area_no_light.csv')
```

```python
df['lat'] = 0.0
df['lon'] = 0.0

for i in range(df.shape[0]):
    latlon = convert_pixel_to_coord(df['px'][i], df['py'][i])
    df['lon'][i] = latlon[0]
    df['lat'][i] = latlon[1]
```
